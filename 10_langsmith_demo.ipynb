```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
```

```python
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system","You are a helpful assistant.Please response to the user request only based on the given context"),
    ("user","Question:{question}\nContext:{context}")]
)
```

```python
model = ChatOpenAI(model="gpt-3.5-turbo")

output_parser = StrOutputParser()
```

```python
chain = prompt | model | output_parser
```

```python
question = "How many colour dragons are there?"

context = """ 
Three dragon brothers—one black, one red, and one green—live on three separate mountains near a small kingdom.
The people in the kingdom are terrified of the dragons, believing they will attack them at any moment.
What the humans see as menacing flying, like spitting flames and chasing each other, is actually just the brothers playing a game of tag high in the sky.
The dragons, in turn, are completely unaware of the fear they are causing, and they have no interest in the kingdom or its inhabitants. 

"""
```

```python
print(chain.invoke({"question":question,"context":context}))
```
